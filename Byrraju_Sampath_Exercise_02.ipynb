## Question 2 (10 Points)
Write Python code to collect a dataset of 1000 samples related to the question discussed in Question 1.

```python
# write your answer here
import random
import pandas as pd
num_samples = 1000
positions = ['Goalkeeper', 'Defender', 'Midfielder', 'Forward']
injury_types = ['None', 'Hamstring', 'Ankle Sprain', 'ACL Tear', 'Concussion', 'Muscle Strain']
def generate_performance_data():
    return {
        'Player_ID': random.randint(1, 100),
        'Age': random.randint(18, 35),
        'Position': random.choice(positions),
        'Distance_Covered_km': round(random.uniform(5, 12), 2),
        'Heart_Rate_BPM': random.randint(60, 180),
        'Injury': random.choices(injury_types, weights=[0.7, 0.1, 0.05, 0.05, 0.05, 0.05])[0],
    }

data = [generate_performance_data() for _ in range(num_samples)]

df = pd.DataFrame(data)

print(df.head(10))
df.to_csv('football_performance_injury_data.csv', index=False)
print("Dataset saved to 'football_performance_injury_data.csv'")
```

## Question 3 (10 Points)
Write Python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "XYZ". The articles should be published in the last 10 years (2014-2024).

The following information from the article needs to be collected:

(1) Title of the article

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
import requests
from bs4 import BeautifulSoup
import csv
import time

keyword = "XYZ"
base_url = "https://dl.acm.org/action/doSearch?AllField={}&AfterYear=2014&BeforeYear=2024&startPage=".format(keyword)

with open('acm_articles.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Venue', 'Year', 'Authors', 'Abstract'])


    for page in range(1, 100):
        url = base_url + str(page)
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')


        articles = soup.find_all('div', class_='issue-item__content')

        for article in articles:
            title_tag = article.find('h5', class_='issue-item__title')
            title = title_tag.get_text(strip=True) if title_tag else 'N/A'
            venue_tag = article.find('span', class_='epub-section__title')
            venue = venue_tag.get_text(strip=True) if venue_tag else 'N/A'
            year_tag = article.find('span', class_='epub-section__date')
            year = year_tag.get_text(strip=True).split()[-1] if year_tag else 'N/A'
            authors_tag = article.find('ul', class_='rlist--inline')
            authors = ', '.join([a.get_text(strip=True) for a in authors_tag.find_all('a')]) if authors_tag else 'N/A'
            abstract_tag = article.find('div', class_='issue-item__abstract')
            abstract = abstract_tag.get_text(strip=True) if abstract_tag else 'N/A'

            writer.writerow([title, venue, year, authors, abstract])
        time.sleep(1)
print("Data collection complete.")
```
